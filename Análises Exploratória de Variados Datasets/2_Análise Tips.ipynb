```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head() # printando somente as cinco primeiras linhas do dataset
```

#### vamos renomear as colunas passando do inglês para o português para melhorar o enrendimento 

```python
tips.columns = ['total conta','gorjeta','sexo','fumante','dia','hora','tamanho']
```

```python
tips.head() # vamos ver como ficou nosso dataset após a renomeação das colunas
```

```python
tips.info() # vamos ver as informaçãoes mas pertinentes, como a quantidade de linhas e se tem algum dado faltante 
```

```python
sns.relplot(x = 'total conta',y = 'gorjeta',hue = 'sexo',style = 'sexo',data = tips) # nesse plot podemos difenciar masculino do feminino
```

```python
sns.relplot(x = 'total conta',y = 'gorjeta',size = 'tamanho',sizes = (15,300),palette = 'blues',data = tips) # nsesse plot fizemos uma relação com a coluna tamanho
```

```python
sns.catplot(x = 'dia',y = 'total conta',hue = 'sexo',data = tips,palette = 'Set1',kind= 'swarm',split = True)
```

#### Graficos de regressão para visualizar a disperção entre as variaveis

```python
sns.lmplot(x = 'total conta',y = 'gorjeta',hue = 'sexo',palette = 'Set2',data = tips) # nesse plot podemos ver que tem uma relação direta com o total da conta 
# com a gorjeta 
```

##### Conclusão: Podemos observar que existe uma relaçaõ entre a total da conta com o a gorjeta paga, quanto mais alto a conta mais alta fica a gorjeta,podemos observar tambem que o sexo feminino frenquenta mais no começo da semana enquanto os homens frequentam mais nos finais de semana.
